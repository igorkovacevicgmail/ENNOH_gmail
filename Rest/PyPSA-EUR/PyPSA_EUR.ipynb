```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load Italian regional boundaries (NUTS2) from a reliable GeoJSON source
regions_url = 'https://raw.githubusercontent.com/openpolis/geojson-italy/master/geojson/limits_IT_regions.geojson'

try:
    italy_regions = gpd.read_file(regions_url)

    # Ensure both datasets use the same Coordinate Reference System (CRS)
    if italy_nodes.crs is None:
        italy_nodes.set_crs(epsg=4326, inplace=True)
    italy_regions = italy_regions.to_crs(italy_nodes.crs)

    # Spatial join: Assign each node to an Italian region
    nodes_with_regions = gpd.sjoin(italy_nodes, italy_regions, predicate='within')

    # Generate the summary table
    # Note: 'reg_name' is the standard column for region names in this GeoJSON
    summary_table = nodes_with_regions.groupby('reg_name').size().reset_index(name='Node Count')
    summary_table = summary_table.sort_values(by='Node Count', ascending=False)

    print("--- Summary of Gas Nodes by Italian Region ---")
    display(summary_table)

    # Quick visualization to verify placement
    fig, ax = plt.subplots(figsize=(10, 10))
    italy_regions.plot(ax=ax, color='whitesmoke', edgecolor='black', linewidth=0.5)
    nodes_with_regions.plot(ax=ax, column='reg_name', markersize=20, legend=True, cmap='tab20')
    plt.title('Distribution of Gas Infrastructure Nodes by Region')
    plt.show()

except Exception as e:
    print(f"An error occurred during regional aggregation: {e}")
    print("Falling back to basic coordinate count if regional data is unreachable.")
```

## Clustering Gas Nodes into NUTS2 Regions

To move toward a sector-coupled model, we often need to aggregate high-resolution nodal data into administrative or market zones (NUTS2). The following code performs this 'clustering' by grouping the spatial join results.

```python
import pandas as pd

# We will use the 'nodes_with_regions' GeoDataFrame created in the previous step
# to 'cluster' the infrastructure metrics.

# For clustering, we usually want the centroids or representative locations of these clusters
# and an aggregate count/capacity.
nuts2_clusters = nodes_with_regions.groupby('reg_name').agg({
    'geometry': lambda x: x.unary_union.centroid,  # Find the geometric center of nodes in that region
    'id': 'count',                                 # Number of facilities
    'name': lambda x: list(x.dropna().unique())[:3] # Sample names of facilities in the cluster
}).rename(columns={'id': 'total_nodes', 'name': 'sample_facilities'})

# Convert back to a GeoDataFrame
nuts2_clusters = gpd.GeoDataFrame(nuts2_clusters, geometry='geometry', crs=nodes_with_regions.crs)

print("--- NUTS2 Clustered Gas Infrastructure Summary ---")
display(nuts2_clusters.sort_values(by='total_nodes', ascending=False))

# Plot the clustered nodes (centroids) over the regional map
fig, ax = plt.subplots(figsize=(10, 10))
italy_regions.plot(ax=ax, color='whitesmoke', edgecolor='black', alpha=0.5)
nuts2_clusters.plot(ax=ax, color='red', markersize=nuts2_clusters['total_nodes'] * 2, alpha=0.7, label='Clustered Centroids')

for x, y, label in zip(nuts2_clusters.geometry.x, nuts2_clusters.geometry.y, nuts2_clusters.index):
    ax.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points", fontsize=8, color='darkblue')

plt.title('Gas Infrastructure Clustered by NUTS2 Region (Centroids scaled by Node Count)')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import pypsa

# 1. Load the regional shapes (if not already in memory)
regions_url = 'https://raw.githubusercontent.com/openpolis/geojson-italy/master/geojson/limits_IT_regions.geojson'
italy_regions = gpd.read_file(regions_url)

# 2. Identify neighbors to create links (Inter-regional NTC)
# We use a spatial overlap/touch approach to find adjacent NUTS2 regions
adjacencies = []
for i, row in italy_regions.iterrows():
    # Find regions that touch the current region
    neighbors = italy_regions[italy_regions.geometry.touches(row.geometry)]
    for j, neighbor in neighbors.iterrows():
        # Avoid double counting (A-B and B-A)
        if row['reg_name'] < neighbor['reg_name']:
            # Assign a proxy NTC value (in MW)
            # In a real model, this would come from Terna's TYNDP or ENTSO-E data
            # Using 1000-3000 MW as a common proxy for inter-regional corridors
            ntc_value = 2000
            adjacencies.append({
                'bus0': row['reg_name'],
                'bus1': neighbor['reg_name'],
                'p_nom': ntc_value
            })

ntc_df = pd.DataFrame(adjacencies)

# 3. Add these to a PyPSA Network as Links
n = pypsa.Network()

# Add Buses for each region
for region in italy_regions['reg_name']:
    n.add("Bus", region, carrier="AC")

# Add Links (NTC) between regions
for i, link in ntc_df.iterrows():
    n.add("Link",
          f"{link['bus0']} - {link['bus1']}",
          bus0=link['bus0'],
          bus1=link['bus1'],
          p_nom=link['p_nom'],
          capital_cost=100, # Example cost
          carrier="electricity")

print(f"--- Created PyPSA Network with {len(n.buses)} Regional Nodes ---")
print(f"--- Added {len(n.links)} Inter-regional NTC Links ---")

# Show sample of the NTC Link table
display(n.links[['bus0', 'bus1', 'p_nom']].head(10))
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

def aggregate_nuts(nodes, url, level_name):
    nuts_gdf = gpd.read_file(url)
    if nodes.crs is None: nodes.set_crs(epsg=4326, inplace=True)
    nuts_gdf = nuts_gdf.to_crs(nodes.crs)

    # Filter for Germany (DE)
    de_nuts = nuts_gdf[nuts_gdf['NUTS_ID'].str.startswith('DE', na=False)]

    # Spatial Join
    nodes_de = gpd.sjoin(nodes, de_nuts, predicate='within')

    # Aggregate
    summary = nodes_de.groupby(['NUTS_ID', 'NAME_LATN']).agg({
        'geometry': lambda x: x.union_all().centroid,
        'id': 'count'
    }).rename(columns={'id': 'node_count'}).reset_index()

    return gpd.GeoDataFrame(summary, geometry='geometry', crs=nodes.crs), de_nuts

urls = {
    'NUTS1': 'https://gisco-services.ec.europa.eu/distribution/v2/nuts/geojson/NUTS_RG_60M_2021_4326_LEVL_1.geojson',
    'NUTS2': 'https://gisco-services.ec.europa.eu/distribution/v2/nuts/geojson/NUTS_RG_60M_2021_4326_LEVL_2.geojson',
    'NUTS3': 'https://gisco-services.ec.europa.eu/distribution/v2/nuts/geojson/NUTS_RG_60M_2021_4326_LEVL_3.geojson'
}

if 'gas_nodes' in globals():
    print('--- Processing Germany NUTS Hierarchy ---')

    for level, url in urls.items():
        cluster, shapes = aggregate_nuts(gas_nodes, url, level)

        fig, ax = plt.subplots(figsize=(10, 8))
        shapes.plot(ax=ax, color='whitesmoke', edgecolor='black', linewidth=0.3)
        cluster.plot(ax=ax, color='teal', markersize=cluster['node_count']*5, alpha=0.7, edgecolor='white')
        plt.title(f'Germany Gas Infrastructure: {level}')
        plt.show()

        print(f'Top 5 {level} Regions in Germany:')
        display(cluster.sort_values(by='node_count', ascending=False).head(5))
else:
    print('Error: gas_nodes not found.')
```

```python
import pypsa
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# 1. Initialize Network
n_italy = pypsa.Network()
n_italy.name = "Italy_NUTS2_Sector_Coupled"

# 2. Add Buses (Electricity and Gas for each NUTS2 region)
for region in italy_regions['reg_name']:
    n_italy.add("Bus", f"{region} AC", carrier="AC")
    n_italy.add("Bus", f"{region} gas", carrier="gas")

# 3. Process Gas Pipelines from IGGIELGNC to NUTS2 Links
# Extract start and end points of Italian pipelines to build connections
if 'italy_pipes' in globals() and not italy_pipes.empty:
    # Create DataFrames for start and end points of the LineStrings
    starts = italy_pipes.copy()
    starts['geometry'] = starts.geometry.apply(lambda geom: Point(geom.coords[0]))

    ends = italy_pipes.copy()
    ends['geometry'] = ends.geometry.apply(lambda geom: Point(geom.coords[-1]))

    # Spatial join with regions to find which region the points fall into
    starts_with_regions = gpd.sjoin(starts, italy_regions[['reg_name', 'geometry']], predicate='within')
    ends_with_regions = gpd.sjoin(ends, italy_regions[['reg_name', 'geometry']], predicate='within')

    # Merge back to find connections between regions
    pipe_links = pd.merge(starts_with_regions[['id', 'reg_name']],
                          ends_with_regions[['id', 'reg_name']],
                          on='id', suffixes=('_start', '_end'))

    # Filter for cross-regional pipes (ignoring internal distribution for the transmission model)
    cross_regional_pipes = pipe_links[pipe_links['reg_name_start'] != pipe_links['reg_name_end']]

    # Aggregate proxy capacities based on pipeline count
    # (Using a standard 5000 MW per pipe segment proxy for this demonstration)
    gas_links_agg = cross_regional_pipes.groupby(['reg_name_start', 'reg_name_end']).size().reset_index(name='pipe_count')
    gas_links_agg['p_nom'] = gas_links_agg['pipe_count'] * 5000

    # Add Gas Links to the network
    for _, row in gas_links_agg.iterrows():
        bus0 = f"{row['reg_name_start']} gas"
        bus1 = f"{row['reg_name_end']} gas"
        # Ensure consistent naming to avoid duplicate bidirectional links
        nodes = sorted([row['reg_name_start'], row['reg_name_end']])
        link_name = f"GasPipe {nodes[0]} - {nodes[1]}"

        if link_name not in n_italy.links.index:
            n_italy.add("Link", link_name,
                        bus0=bus0, bus1=bus1,
                        p_nom=row['p_nom'],
                        carrier="gas",
                        bidirectional=True)
    print(f"Added {len(gas_links_agg)} aggregated cross-regional gas pipeline links.")
else:
    print("italy_pipes dataset not found. Skipping physical gas network links.")

# 4. Add Sector Coupling (Gas to Power)
for region in italy_regions['reg_name']:
    n_italy.add("Link", f"{region} CCGT",
                bus0=f"{region} gas", bus1=f"{region} AC",
                p_nom=5000, efficiency=0.6, carrier="gas-to-power")

# 5. Add LNG Terminals
lng_terminals = [
    {"name": "Panigaglia LNG", "bus": "Liguria gas", "p_nom": 10000},
    {"name": "OLT Livorno LNG", "bus": "Toscana gas", "p_nom": 12000},
    {"name": "Porto Viro LNG", "bus": "Veneto gas", "p_nom": 25000}
]
for lng in lng_terminals:
    n_italy.add("Generator", lng["name"], bus=lng["bus"], p_nom=lng["p_nom"], carrier="gas", marginal_cost=50)

print("\n--- PyPSA NUTS2 Italy Model Built ---")
print(f"Total Buses: {len(n_italy.buses)}")
print(f"Total Links: {len(n_italy.links)}")
print(f"Total Generators: {len(n_italy.generators)}")

# Display the physical cross-regional gas pipelines
display(n_italy.links[n_italy.links.carrier == 'gas'][['bus0', 'bus1', 'p_nom']].head(10))
```

# Europena Parts, National Levels and Regional Levels

```python
import pypsa
import pandas as pd

n = pypsa.Network()

# 1. Define the Mapping
# Countries mapped to 5 Macro-Regions (e.g., North, South, West, East, Central)
mapping = {
    'DE': 'Central', 'FR': 'West', 'BE': 'West', 'NL': 'West',
    'ES': 'South', 'IT': 'South', 'PL': 'East', 'NO': 'North'
}

# 2. Add Country-Level Buses (Level 1)
for country in mapping.keys():
    n.add("Bus", country, carrier="H2")

# 3. Add Regional-Level "Super-Buses" (Level 2)
regions = list(set(mapping.values()))
for region in regions:
    n.add("Bus", f"Region_{region}", carrier="H2_aggregated")

# 4. Connect Countries to their Regional Hub
# We use Links with high capacity and 100% efficiency to act as "Virtual Bridges"
for country, region in mapping.items():
    n.add("Link",
          f"connector_{country}",
          bus0=country,
          bus1=f"Region_{region}",
          p_nom=1e6, # Effectively unconstrained
          efficiency=1.0)

# 5. Place Storages ONLY on the Regional Hubs
for region in regions:
    n.add("Store",
          f"H2_Storage_{region}",
          bus=f"Region_{region}",
          e_nom_extendable=True,
          capital_cost=3000, # Cost per MWh
          carrier="H2_storage")

# 6. Add the European Backbone (Links between Countries)
# Example: DE-FR Link
n.add("Link", "DE-FR", bus0="DE", bus1="FR", p_nom=5000)
```

### Applying Multi-Level Constraints
In our hierarchical model, we can treat country connectors as individual constraints and regional hubs as aggregate constraints.

```python
import pypsa

# 1. Country-Level Constraint: Limit Germany's export/import capacity to the hub
n.links.at['connector_DE', 'p_nom_max'] = 50000 # Limit to 50 GW

# 2. Regional-Level Constraint: Limit total regional storage expansion
# We can set an upper bound on the extendable storage capacity for a specific region
n.stores.at['H2_Storage_Central', 'e_nom_max'] = 5000000 # Max 5 TWh for the Central region

# 3. Check the updated constraints
print("--- Updated Country Connector Limit (DE) ---")
print(n.links.loc[['connector_DE'], ['p_nom', 'p_nom_max']])

print("\n--- Updated Regional Storage Limit (Central) ---")
print(n.stores.loc[['H2_Storage_Central'], ['e_nom_extendable', 'e_nom_max']])
```

### Hierarchical Heat Modeling: State-Level Heat vs. National Electricity
In this setup:
1. **State Level**: Heat demand and supply (District Heating) are managed per German state.
2. **National Level**: The electricity bus `DE` provides the energy for Power-to-Heat (P2H) across all states.
